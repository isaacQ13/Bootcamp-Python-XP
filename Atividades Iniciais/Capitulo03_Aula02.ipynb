## Capitulo 3, Aula 2 - Operadores dos tipos primitivos de dados.

- Operadores Aritméticos (`+`, `-`, `*`, `/`, `%`, `//`, `**`)
- Operadores de Comparação(`>`, `>=`, `<=`, `<`, `==`, `!=`)
- Operadores lógicos(`and`, `or`, `not`)

### Operadores Aritméticos

```python
#Operador De soma
20+20
```

```python
#Operador de subtração
20-20
```

```python
#Operador de multiplicação
20*20
```

```python
#Operador de Divisão
20/3
```

```python
#Operador de Divisão (Retorna numeros inteiros)
20//3
```

```python
#Operador Resto de Divisão
print(20%20)
print(20%3)
```

```python
#Operador de Potenciação
print(2**2)
print(2**8)
```

```python
x = 10
y = 4
z = 4.5
print('x + y + z=', x + y + z)
print('x - 3 =', x - 3)
print('x * z =', x * z)
print('x * 2 =', x * 2)
print('x / (y + 2) =', x / (y + 2))
print('x % y =', x % y)
print('x // y =', x // y)
print('x ** z =', x ** z)
```

### Operadores de Comparação

```python
#Operador Maior Que
print(10>4)
print(4>10)
print(4>4)
```

```python
#Operador Maior ou Igual
print(10>4)
print(4>10)
print(4>=4)
```

```python
#Operador Menor Que
print(10<4)
print(4<10)
```

```python
#Operador Menor ou Igual
print(10<=4)
print(4<=10)
print(4<=4)
```

```python
#Operador de "igualdade"
print(10==4)
print(4==10)
print(4==4)
```

```python
#Operador de Diferença
print(10!=4)
print(4!=10)
print(4!=4)
```
